```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d
from tqdm import tqdm
```

```python
def CreateSphere(N,R=1) -> int:
    
    Points = np.zeros((N,3))
    
    for i in tqdm(range(N)):
        
        phi = np.random.uniform(0,2*np.pi)
        costheta = np.random.uniform(-1,1)
        theta = np.arccos(costheta)
        
        u = np.random.rand()
        
        radio = R*u**(1/3.)
        
        Points[i] = [radio*np.cos(phi)*np.sin(theta),radio*np.sin(phi)*np.sin(theta),radio*np.cos(theta)]

    return Points
```

```python
Esfera = CreateSphere(int(4e5))
```
