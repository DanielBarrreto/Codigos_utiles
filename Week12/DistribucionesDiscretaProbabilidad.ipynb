```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import bernoulli
from scipy.stats import binom
from scipy.stats import poisson
```

```python
def GetHisto(Data,bines):
    
    h, bins = np.histogram(Data,bins=bines)
    w = np.diff(bins)   # Ancho de clase
    m = bins[:-1] + 0.5*w[0] # Marca de clase
    
    hnorm = h/np.sum(h*w)
    
    return h,hnorm,m,w
```

```python
# Bernoulli
# Probabilidad de exito
p = 1/6.
N = 10000
```

```python
b = bernoulli.rvs( p, size=N )
xbins = np.linspace(0,2,3)
Hb,HNb,mb,w = GetHisto(b,xbins)
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)

ax.bar(mb,HNb, width=w, ec='k')
ax.set_ylim(0,1)

for i in range(len(mb)):
    ax.text( mb[i], 0.02 + np.round( HNb[i],2 ), \
             str( np.round( HNb[i]*w[i]*100, 2 ))+'%', ha='center', fontsize=15)
```

```python
#Valor medio
b.mean()
```

```python
# Usando histo
X = np.array([0.,1.])
np.sum( HNb*X )
```

```python
# Binomial
# Probabilidad de obtener dos 2 en un lanzamiento
n, p = 2, 1./6 # Numero de lanzamientos
binomial = binom.rvs(n,p,size=N)
xbins = np.linspace(0,3,4)
Hb,HNb,mb,w = GetHisto(binomial,xbins)
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)

ax.bar(mb,HNb, width=w, ec='k')
ax.set_ylim(0,1)

for i in range(len(mb)):
    ax.text( mb[i], 0.02 + np.round( HNb[i],2 ), \
             str( np.round( HNb[i]*w[i]*100, 2 ))+'%', ha='center', fontsize=15)
```

```python
#Valor medio
binomial.mean()
```

```python
# Usando histo
X = np.array([0.,1.,2.])
np.sum( HNb*X )
```

```python
# Poisson
mu = 0.5
Pois = poisson.rvs( mu, size=N)
xbins = np.linspace(0,6,7)
Hb,HNb,mb,w = GetHisto(Pois,xbins)
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)

ax.bar(mb,HNb, width=w, ec='k')
ax.set_ylim(0,1)

for i in range(len(mb)):
    ax.text( mb[i], 0.02 + np.round( HNb[i],2 ), \
             str( np.round( HNb[i]*w[i]*100, 2 ))+'%', ha='center', fontsize=15)
```

```python
#Valor medio
Pois.mean()
```

```python
# Usando histo
X = np.array([0.,1.,2.,3.,4.,5.])
np.sum( HNb*X )
```
