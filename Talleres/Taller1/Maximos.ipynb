```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os.path as path
import os
import wget
```

```python
if not path.exists('Data'):
    os.mkdir('Data')
    
file = 'Data/Espectro.txt'
url = 'https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/EstrellaEspectro.txt'

if not path.exists(file):
    Path_ = wget.download(url,file)
    
else:
    Path_ = file
```

```python
Data = np.loadtxt(Path_)
#Data
```

```python
x = Data[:,0]
y = Data[:,1]
```

```python
plt.plot(x,y)
```

```python
maxima = []
imagenes = []

for i in range(1,len(x)-1):
    if y[i] > y[i-1] and y[i] > y[i+1]:
        maxima.append(x[i])
        imagenes.append(y[i])
```

```python
plt.plot(x,y)

for i in range(len(maxima)):
    plt.scatter(maxima,imagenes,color='r',marker='.')
    #plt.axvline(x = maxima[i],ls='--',color='r')
    
#plt.savefig('Maximos.pdf')
```
